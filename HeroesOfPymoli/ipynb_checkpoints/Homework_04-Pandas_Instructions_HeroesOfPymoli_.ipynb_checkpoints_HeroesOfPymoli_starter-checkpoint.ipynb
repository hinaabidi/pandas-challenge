```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
df=purchase_data[["Age", "Gender", "SN"]].drop_duplicates()
df.head()
```

```python
num=df.count()[0]
num
```

```python
df_num=pd.DataFrame({"Total Players": [num]})
df_num
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items=purchase_data['Item Name'].nunique()
print(unique_items)
avg_price=purchase_data['Price'].mean()
print(avg_price)
num_purch=purchase_data['Purchase ID'].count()
print(num_purch)
total_revenue=purchase_data['Price'].sum()
total_revenue
summary_table=pd.DataFrame({"Number of Unique Items": [unique_items], 
                           "Average Price": avg_price,
                           "Number of Purchase": num_purch,
                           "Total Revenue": total_revenue})



summary_table["Average Price"]=summary_table["Average Price"].map("${:.2f}".format)
summary_table["Total Revenue"]=summary_table["Total Revenue"].map("${:,.2f}".format)
summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
by_gender=df['Gender'].value_counts()

perc=by_gender/num
perc_df=pd.DataFrame({"Total Count": by_gender, "Percentage of Players": perc})
perc_df['Percentage of Players']=perc_df["Percentage of Players"].map("{:,.2%}".format)
perc_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purch_count=purchase_data.groupby(["Gender"]).count()['SN']
print(purch_count)
avg_price=purchase_data.groupby(["Gender"]).mean()["Price"]
print(avg_price)
total_purch_value=purchase_data.groupby(["Gender"]).sum()["Price"]
total_purch_value
avg_purch_person=total_purch_value/perc_df["Total Count"]
avg_purch_person
summary=pd.DataFrame({"Purchase Count": purch_count,
                     "Average Purchase Price": avg_price,
                     "Total Purchase Value": total_purch_value,
                     "Avg Total Purch per Person": avg_purch_person})
summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9, 9999]
labels=['<10','10-14','15-19', '20-24','25-29','30-34','35-39','40+']
df["Ranges"]=pd.cut(df["Age"], bins, labels=labels)
total_count=df['Ranges'].value_counts()
# print(total_count)
Per_of_pla=total_count/num
# print(Per_of_pla)
new_df=pd.DataFrame({"Total Count":total_count,
                    "Percentage of Players": Per_of_pla})
new_df.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins=[0,9.9,14.9,19.9,24.9,29.9,34.9,39.9, 9999]
labels=['<10','10-14','15-19', '20-24','25-29','30-34','35-39','40+']
purchase_data["Ranges"]=pd.cut(purchase_data["Age"], bins, labels=labels)
Purchase_Count=purchase_data["Ranges"].value_counts()
Purchase_Count
Avg_price=purchase_data.groupby(["Ranges"]).mean()["Price"]
Avg_price
Total_Purch_Value=purchase_data.groupby(["Ranges"]).sum()["Price"]
Total_Purch_Value
Avg_total_per_person=Total_Purch_Value/new_df["Total Count"]
Avg_total_per_person
summary=pd.DataFrame({"Purchase_Count":Purchase_Count,
                     "Average Purchase Price": Avg_price,
                     "Total Purchase Value": Total_Purch_Value,
                     "Avg Total Purchase per Person": Avg_total_per_person})
summary.sort_index()
```
